

Le processus d'Ornstein-Uhlenbeck est défini par l'équation différentielle stochastique (EDS) :

$$
dX_t = \theta (\mu - X_t) dt + \sigma dW_t,
$$

où :

$\theta > 0$ est le paramètre de vitesse de retour à la moyenne,
$\mu$ est la moyenne de long terme,
$\sigma$ est la volatilité,
$W_t$ est un processus de Wiener standard.


On peut réécrire l'EDS sous la forme :

$$
dX_t + \theta X_t dt = \theta \mu dt + \sigma dW_t.
$$

On applique le facteur d'intégration $e^{\theta t}$ pour la formule d'ito que l'on applique à $f(X_t,t) = X_t e^{\theta  t} $:

$$
e^{\theta t} dX_t + \theta e^{\theta t} X_t dt = \theta \mu e^{\theta t} dt + \sigma e^{\theta t} dW_t.
$$

En intégrant de $0$ à $t$, on obtient :

$$
X_t e^{\theta t} - X_0 = \theta \mu \int_0^t e^{\theta s} ds + \sigma \int_0^t e^{\theta s} dW_s.
$$

On calcule l'intégrale déterministe :

$$
\int_0^t e^{\theta s} ds = \frac{e^{\theta t} - 1}{\theta}.
$$

On en déduit la solution :

$$
X_t = X_0 e^{-\theta t} + \mu(1 - e^{-\theta t}) + \sigma \int_0^t e^{-\theta (t-s)} dW_s.
$$


La solution du processus d'Ornstein-Uhlenbeck à un instant $t+\Delta t $ est donnée par :

$$
X_{t+\Delta t} = X_0 e^{-\theta (t+\Delta t)} + \mu(1 - e^{-\theta (t+\Delta t)}) + \sigma \int_0^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s.
$$

On décompose l'intégrale en deux parties :

$$
\int_0^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s = \int_0^t e^{-\theta (t+\Delta t -s)} dW_s + \int_t^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s.
$$

Ainsi, on peut réécrire $ X_{t+\Delta t} $ en séparant les termes dépendant de $ X_t $ et ceux qui en sont indépendants :

$$
X_{t+\Delta t} = X_0 e^{-\theta t} e^{-\theta \Delta t} + \mu(1 - e^{-\theta (t+\Delta t)}) + \sigma \int_0^t e^{-\theta (t+\Delta t -s)} dW_s + \sigma \int_t^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s.
$$

Or, de l’équation de $ X_t $, on sait que :

$$
X_t = X_0 e^{-\theta t} + \mu(1 - e^{-\theta t}) + \sigma \int_0^t e^{-\theta (t-s)} dW_s.
$$

En multipliant cette expression par $ e^{-\theta \Delta t} $, on obtient :

$$
X_t e^{-\theta \Delta t} = X_0 e^{-\theta t} e^{-\theta \Delta t} + \mu(1 - e^{-\theta t}) e^{-\theta \Delta t} + \sigma \int_0^t e^{-\theta (t-s)} e^{-\theta \Delta t} dW_s.
$$

En réarrangeant, on trouve :

$$
X_{t+\Delta t} = X_t e^{-\theta \Delta t} + \mu(1 - e^{-\theta (t+\Delta t)}) + \sigma \int_t^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s.
$$


L'espérance conditionnelle $ \mathbb{E}[X_{t+\Delta t} | X_t] $ est obtenue en prenant l'espérance des deux côtés :

$$
\mathbb{E}[X_{t+\Delta t} | X_t] = \mathbb{E}[X_t e^{-\theta \Delta t} + \mu(1 - e^{-\theta (t+\Delta t)}) + \sigma \int_t^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s | X_t].
$$

On utilise la propriété que l’espérance d’une intégrale stochastique est nulle :

$$
\mathbb{E} \left[ \int_t^{t+\Delta t} e^{-\theta (t+\Delta t -s)} dW_s \Big| X_t \right] = 0.
$$

Ainsi, il reste :

$$
\mathbb{E}[X_{t+\Delta t} | X_t] = X_t e^{-\theta \Delta t} + \mu(1 - e^{-\theta (t+\Delta t)}).
$$

Comme $ t $ est arbitraire et que nous nous intéressons uniquement à l’évolution sur $ \Delta t $, on peut réécrire :

$$
\mathbb{E}[X_{t+\Delta t} | X_t] = X_t e^{-\theta \Delta t} + \mu(1 - e^{-\theta \Delta t}).
$$


Cela montre que le processus suit un mouvement exponentiel vers $\mu$.


La variance est obtenue en calculant la variance du terme aléatoire :

$$
\text{Var}(X_{t+\Delta t} | X_t = x) = \mathbb{E} \left[ \left( \sigma \int_t^{t+\Delta t} e^{-\theta (t+\Delta t - s)} dW_s \right)^2 \right].
$$

Utilisant l'isométrie d'Itô :

$$
\mathbb{E} \left[ \left( \int_a^b f(s) dW_s \right)^2 \right] = \int_a^b f^2(s) ds,
$$

on obtient :

$$
\text{Var}(X_{t+\Delta t} | X_t = x) = \sigma^2 \int_t^{t+\Delta t} e^{-2\theta (t+\Delta t - s)} ds.
$$

En intégrant :

$$
\int_t^{t+\Delta t} e^{-2\theta (t+\Delta t - s)} ds = \frac{1 - e^{-2\theta \Delta t}}{2\theta}.
$$

Finalement, on obtient la variance conditionnelle :

$$
\sigma_t^2 = \frac{\sigma^2}{2\theta} (1 - e^{-2\theta \Delta t}).
$$




La probabilité de transition $ P[i,j] $ est définie comme :

$
P[i,j] = \mathbb{P}(X_{t+\Delta t} \in I_j \mid X_t \in I_i)
$

où $ I_i = [x_i - \frac{\delta x}{2}, x_i + \frac{\delta x}{2}] $ et $ I_j = [x_j - \frac{\delta x}{2}, x_j + \frac{\delta x}{2}])$.  
Étant donné que $ X_{t+\Delta t} \mid X_t = x $ suit une loi normale de moyenne :
$$
m(x) = x e^{-\theta \Delta t} + \mu (1 - e^{-\theta \Delta t})
$$

et de variance :

$$
\sigma_t^2 = \frac{\sigma^2}{2\theta} (1 - e^{-2\theta \Delta t}),
$$

la probabilité de transition est obtenue en intégrant la densité normale sur $I_j$, en tenant compte de la distribution uniforme de $X_t$ sur $I_i$ :

$$
P[i,j] = \int_{I_i} \left( \int_{I_j} \frac{1}{\sqrt{2\pi \sigma_t^2}} \exp \left( -\frac{(y - m(x))^2}{2\sigma_t^2} \right) dy \right) \frac{dx}{\delta x}
$$

Cette intégrale double exprime la probabilité qu'une valeur initiale $X_t$ dans  $I_i$ conduise à une valeur $X_{t+\Delta t}$ dans $I_j$, en tenant compte de la dynamique du processus d'Ornstein-Uhlenbeck.


```python
import numpy as np
import scipy.stats as stats
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import seaborn as sns

def probabilite_transition(x_i, x_j, dx, theta, mu, sigma, T, x_min, x_max):
    def integrande(x, y):
        moyenne_t = x * np.exp(-theta * T) + mu * (1 - np.exp(-theta * T))
        ecart_type_t = np.sqrt(sigma**2 / (2 * theta) * (1 - np.exp(-2 * theta * T)))
        return (stats.norm.pdf(y, loc=moyenne_t, scale=ecart_type_t) / dx) * (1 / dx)

    # Gérer les conditions aux limites
    borne_inf_x = -np.inf if x_i == x_min else x_i - dx/2
    borne_sup_x = np.inf if x_i == x_max else x_i + dx/2

    probabilite, _ = integrate.dblquad(integrande, borne_inf_x, borne_sup_x, lambda x: x_j - dx/2, lambda x: x_j + dx/2)
    return probabilite

def matrice_transition_ornstein_uhlenbeck(theta, mu, sigma, dt, x_min, x_max, n_intervalles):
    valeurs_x = np.linspace(x_min, x_max, n_intervalles)  # Centres des intervalles
    dx = (x_max - x_min) / (n_intervalles - 1)  # Largeur d'un intervalle
    matrice_P = np.zeros((n_intervalles, n_intervalles))  # Matrice de transition

    # Calcul de la matrice de transition avec intégration
    for i in range(n_intervalles):
        for j in range(n_intervalles):
            matrice_P[i, j] = probabilite_transition(valeurs_x[i], valeurs_x[j], dx, theta, mu, sigma, T, x_min, x_max)
        matrice_P[i, :] /= matrice_P[i, :].sum()

    return valeurs_x, matrice_P

def afficher_matrice_transition(valeurs_x, matrice_P):
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrice_P, xticklabels=np.round(valeurs_x, 2), yticklabels=np.round(valeurs_x, 2), cmap="viridis", annot=False)
    plt.xlabel("État à t+dt")
    plt.ylabel("État à t")
    plt.title("Matrice de transition du processus d'Ornstein-Uhlenbeck")
    plt.show()

# Paramètres du processus OU
theta = 1.0  # Intensité du retour à la moyenne
mu = 0.0  # Moyenne de long terme
sigma = 0.5  # Volatilité
T = 1  #  temps
x_min, x_max = -2, 2  # Bornes de l'espace d'état
n_intervalles = 20  # Nombre d'intervalles

# Génération et affichage de la matrice de transition
valeurs_x, matrice_P = matrice_transition_ornstein_uhlenbeck(theta, mu, sigma, T, x_min, x_max, n_intervalles)
afficher_matrice_transition(valeurs_x, matrice_P)
```
